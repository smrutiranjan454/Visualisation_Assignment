```python
#1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset.
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(0)  
x = np.random.rand(100) * 100  
y = 2.5 * x + np.random.randn(100) * 10  

data = pd.DataFrame({'x': x, 'y': y})

sns.scatterplot(data=data, x='x', y='y', color='blue')

plt.title('Scatter Plot of Synthetic Data')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')

plt.show()

```

```python
#2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(0)  
data = np.random.randn(1000)  

df = pd.DataFrame({'value': data})

sns.histplot(df['value'], bins=30, kde=True, color='skyblue')

plt.title('Distribution of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```

```python
#3. Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value': [25, 40, 15, 30, 20]
}

df = pd.DataFrame(data)

sns.barplot(data=df, x='Category', y='Value', palette='viridis')

plt.title('Comparison of Categories Based on Values')
plt.xlabel('Category')
plt.ylabel('Value')

plt.show()
```

```python
#4.Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

np.random.seed(0)  
categories = ['A', 'B', 'C', 'D', 'E']
data = {
    'Category': np.random.choice(categories, size=1000),
    'Value': np.random.randn(1000) * 10 + 50  
}

df = pd.DataFrame(data)

sns.boxplot(data=df, x='Category', y='Value', palette='Set2')

plt.title('Distribution of Values Across Categories')
plt.xlabel('Category')
plt.ylabel('Value')

plt.show()
```

```python
#5.Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(0)  
n = 100  
x = np.random.randn(n)
y = 2 * x + np.random.randn(n) * 0.5  
z = -0.5 * x + np.random.randn(n) * 0.5


df = pd.DataFrame({
    'Feature 1': x,
    'Feature 2': y,
    'Feature 3': z
})

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

plt.title('Correlation Matrix Heatmap')

plt.show()
```
