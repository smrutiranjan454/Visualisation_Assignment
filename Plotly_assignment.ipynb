```python
#3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
"""
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
} 
df = pd.DataFrame(data)
"""
import plotly.express as px
import numpy as np
import pandas as pd

np.random.seed(20)

data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

heatmap_data = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

fig = px.imshow(heatmap_data, color_continuous_scale='Viridis', title='Variation in Sales Across Months and Days')

fig.update_layout(
    xaxis_title='Month',
    yaxis_title='Day',
    xaxis=dict(tickmode='array', tickvals=list(heatmap_data.columns), ticktext=list(heatmap_data.columns)),
    yaxis=dict(tickmode='array', tickvals=list(heatmap_data.index), ticktext=list(heatmap_data.index))
)
fig.show()
```

```python
#4. Using the given x and y data, generate a 3D surface plot to visualize the function .
"""
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2)) 
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
} 
df = pd.DataFrame(data)
"""
import plotly.graph_objects as go
import numpy as np
import pandas as pd

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

fig = go.Figure(data=[go.Surface(
    z=df.pivot('Y', 'X', 'Z').values,
    x=df.pivot('Y', 'X', 'Z').columns,
    y=df.pivot('Y', 'X', 'Z').index
)])

fig.update_layout(
    title='3D Surface Plot of the Function z = sin(sqrt(x^2 + y^2))',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    )
)

fig.show()
```
